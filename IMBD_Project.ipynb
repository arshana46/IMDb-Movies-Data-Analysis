```python
df['Metascore']=df['Metascore'].fillna(df['Metascore'].mean())
df

                                       
```

```python

df['age_limit'] = df['age_limit'].fillna(df['age_limit'].mode()[0])

df.isnull().sum()
```

```python
import re

def convert_duration(x):
    if isinstance(x, str):
        match = re.match(r'(?:(\d+)h)?\s*(?:(\d+)m)?', x)
        if match:
            hours = int(match.group(1)) if match.group(1) else 0
            minutes = int(match.group(2)) if match.group(2) else 0
            return hours * 60 + minutes
    return None  # for missing or invalid values

df['duration_minutes'] = df['duration'].apply(convert_duration)
df.head()
```

```python
plt.figure(figsize=(12,8))
plt.hist(df['rating'],bins=10,edgecolor='Black')
plt.title('Distribution of imbd ratings')
plt.xlabel('imdb rating')
plt.ylabel('number of movies')
plt.show()
         
```

```python
highest = df.loc[df['rating'].idxmax()]
lowest = df.loc[df['rating'].idxmin()]

print("🎬 Highest Rated Movie:")
print(highest[['name', 'rating', 'year']])
print("\n🎞️ Lowest Rated Movie:")
print(lowest[['name', 'rating', 'year']])
```

```python
df.groupby('year')['rating'].mean().plot(kind='line', figsize=(10,5))
plt.title('Average IMDb Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.scatter(df['duration_minutes'],df['rating'])
plt.title('Movie Duration vs IMDb Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
corr=df[['rating', 'Metascore', 'duration_minutes']].corr()
corr
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for IMDb Data')
plt.show()

# +1 → strong positive relationship

# -1 → strong negative relationship

# 0 → no relationship

# 🔵 Blue = positive correlation

# 🔴 Red = negative correlation

# ⚪ Near white = weak or no correlation
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Metascore', y='rating', data=df)
plt.title('Metascore vs IMDb Rating')
plt.xlabel('Metascore (Critic)')
plt.ylabel('IMDb Rating (Audience)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby('age_limit')['rating'].mean().sort_values().plot(kind='bar', color='purple')
plt.title('Average Rating by Age Limit')
plt.xlabel('Age Limit')
plt.ylabel('Average Rating')
plt.show()
```

## IMDb Movies Data Analysis using Python

#### 🎯 Objective:

The goal of this project is to analyze IMDb movie data to understand trends in ratings, duration, metascore, and other movie features using Python libraries like Pandas, NumPy, Matplotlib, and Seaborn.

#### 🧹 Data Cleaning:

Removed or filled missing values.

Converted duration from "2h 22m" → 142 minutes.

Checked and removed duplicates.

Verified no null values after cleaning.

#### 🔍 Data Exploration & Analysis:

Ratings Distribution:
Most movies are rated between 7.5 and 8.0, showing that the dataset mainly includes high-quality movies.

Ratings by Year:
Movie ratings vary across years, with older classics often maintaining higher ratings.

Duration vs Rating:
No strong relationship — both short and long movies can have high ratings.

Metascore vs Rating:
Moderate positive correlation (~0.6–0.8). Critics and audience opinions usually align.

Age Limit vs Rating:
Movies for older audiences (e.g., 18+) tend to have higher ratings.

#### 📊 Correlation Heatmap:

Showed positive correlations between metascore and rating, indicating consistent opinions between critics and audiences.

#### 🧠 Key Insights:

Longer duration doesn’t guarantee higher ratings.

Critics and audiences generally agree, but not always.

Older films maintain strong reputations over time.

18+ or adult-rated movies tend to have better storytelling and ratings.
